## Logistic Regression

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import KFold
import numpy as np

def logistic():
    clf = LogisticRegression(C=0.2, solver='liblinear', max_iter=10000)
    return clf

# Initialize lists to store results
lr_accuracy = []
lr_precision = []
lr_recall = []
lr_f1 = []
lr_fpr = []

# For tracking accuracy per fold across all attack types
fold_accuracies = [[] for _ in range(5)]  # 5 folds

kf = KFold(n_splits=5, shuffle=True, random_state=42)

for attack_type in attack_classes:
    print(f"\n=== Attack Type: {attack_type} ===")

    acc_folds = []
    precision_folds = []
    recall_folds = []
    f1_folds = []
    fpr_folds = []

    for fold_idx, (train_index, test_index) in enumerate(kf.split(x_train), 1):
        x_train_fold, x_test_fold = x_train.iloc[train_index], x_train.iloc[test_index]
        y_train_fold, y_test_fold = train_label[attack_type].iloc[train_index], train_label[attack_type].iloc[test_index]

        clf = logistic()
        clf.fit(x_train_fold, y_train_fold)
        y_pred = clf.predict(x_test_fold)

        acc = accuracy_score(y_test_fold, y_pred)
        precision = precision_score(y_test_fold, y_pred, zero_division=0)
        recall = recall_score(y_test_fold, y_pred, zero_division=0)
        f1 = f1_score(y_test_fold, y_pred, zero_division=0)

        cm = confusion_matrix(y_test_fold, y_pred)
        if cm.shape == (2, 2):
            tn, fp, fn, tp = cm.ravel()
            fpr = fp / (fp + tn) if (fp + tn) > 0 else 0
        else:
            fpr = 0

        acc_folds.append(acc)
        precision_folds.append(precision)
        recall_folds.append(recall)
        f1_folds.append(f1)
        fpr_folds.append(fpr)

        # Collect fold-wise accuracies across all attack types
        fold_accuracies[fold_idx - 1].append(acc)

        print(f"Fold {fold_idx}: Accuracy = {acc * 100:.2f}%")

    mean_acc = np.mean(acc_folds)
    mean_precision = np.mean(precision_folds)
    mean_recall = np.mean(recall_folds)
    mean_f1 = np.mean(f1_folds)
    mean_fpr = np.mean(fpr_folds)

    lr_accuracy.append(mean_acc)
    lr_precision.append(mean_precision)
    lr_recall.append(mean_recall)
    lr_f1.append(mean_f1)
    lr_fpr.append(mean_fpr)

    print(f"Average Accuracy for {attack_type}: {mean_acc * 100:.2f}%")

# After all attack types, compute and print fold averages
print("\n=== Average Accuracy per Fold across all Attack Types ===")
for i, fold_acc_list in enumerate(fold_accuracies, 1):
    avg_fold_acc = np.mean(fold_acc_list)
    print(f"Fold {i} Average Accuracy: {avg_fold_acc * 100:.2f}%")

# Overall mean across all folds and attack types
print("\n=== Logistic Regression Classifier - Overall Results ===")
print(f"Overall Mean Accuracy    : {np.mean(lr_accuracy) * 100:.2f}%")
```

## Stacked Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization

# Logistic Regression classifier
def logistic():
    clf = LogisticRegression(C=0.2, solver='liblinear', max_iter=10000)
    return clf

# SGD classifier
def get_sgd_classifier():
    return SGDClassifier(max_iter=1000, tol=1e-3, learning_rate='optimal')

# Random Forest classifier
def get_rfc_classifier():
    return RandomForestClassifier()

# DNN classifier
def get_dnn_classifier(input_dim=41):
    clf = Sequential()
    clf.add(Dense(1024, input_dim=input_dim, activation='relu'))
    clf.add(BatchNormalization())
    clf.add(Dense(1024, activation='relu'))
    clf.add(BatchNormalization())
    clf.add(Dense(512, activation='relu'))
    clf.add(Dense(64, activation='relu'))
    clf.add(Dense(1, activation='sigmoid'))
    clf.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return clf

# Stacking predictions
stacked_train = {attack_type: {'logistic': [], 'sgd': [], 'rfc': [], 'dnn': []} for attack_type in attack_classes}
stacked_test = {attack_type: {'logistic': [], 'sgd': [], 'rfc': [], 'dnn': []} for attack_type in attack_classes}

# Initialize lists for storing the accuracy results
fold_accuracies = [[] for _ in range(5)]  # For storing fold-wise accuracies

kf = KFold(n_splits=5, shuffle=True, random_state=42)

fold_num = 1
for train_index, val_index in kf.split(x_train):
    print(f"Fold {fold_num}...")

    # Create the training and validation splits
    x_train_fold, x_val_fold = x_train.iloc[train_index], x_train.iloc[val_index]
    y_train_fold, y_val_fold = train_label[attack_type].iloc[train_index], train_label[attack_type].iloc[val_index]

    # Train each model
    clf_logistic = logistic()
    clf_sgd = get_sgd_classifier()
    clf_rfc = get_rfc_classifier()
    clf_dnn = get_dnn_classifier(input_dim=x_train.shape[1])

    clf_logistic.fit(x_train_fold, y_train_fold)
    clf_sgd.fit(x_train_fold, y_train_fold)
    clf_rfc.fit(x_train_fold, y_train_fold)
    clf_dnn.fit(x_train_fold, y_train_fold, batch_size=1024, epochs=3, verbose=0, validation_data=(x_val_fold, y_val_fold))

    # Predictions for stacking (train and validation sets)
    logistic_train_preds = clf_logistic.predict(x_train_fold)
    sgd_train_preds = clf_sgd.predict(x_train_fold)
    rfc_train_preds = clf_rfc.predict(x_train_fold)
    dnn_train_preds = (clf_dnn.predict(x_train_fold) >= 0.5).astype(int)

    logistic_val_preds = clf_logistic.predict(x_val_fold)
    sgd_val_preds = clf_sgd.predict(x_val_fold)
    rfc_val_preds = clf_rfc.predict(x_val_fold)
    dnn_val_preds = (clf_dnn.predict(x_val_fold) >= 0.5).astype(int)

    # Store predictions for the stacked model
    stacked_train[attack_type]['logistic'].append(logistic_train_preds)
    stacked_train[attack_type]['sgd'].append(sgd_train_preds)
    stacked_train[attack_type]['rfc'].append(rfc_train_preds)
    stacked_train[attack_type]['dnn'].append(dnn_train_preds)

    stacked_test[attack_type]['logistic'].append(logistic_val_preds)
    stacked_test[attack_type]['sgd'].append(sgd_val_preds)
    stacked_test[attack_type]['rfc'].append(rfc_val_preds)
    stacked_test[attack_type]['dnn'].append(dnn_val_preds)

    # Stack predictions for training the logistic regression model
    X_train_stack = np.column_stack((logistic_train_preds, sgd_train_preds, rfc_train_preds, dnn_train_preds))
    X_val_stack = np.column_stack((logistic_val_preds, sgd_val_preds, rfc_val_preds, dnn_val_preds))

    # Train Logistic Regression classifier on the stacked features
    clf_logistic_stack = LogisticRegression(C=0.2, solver='liblinear', max_iter=10000)
    clf_logistic_stack.fit(X_train_stack, y_train_fold)

    # Make final predictions
    final_preds = clf_logistic_stack.predict(X_val_stack)
    fold_accuracy = accuracy_score(y_val_fold, final_preds)

    fold_accuracies[fold_num - 1].append(fold_accuracy)

    print(f"Accuracy for Fold {fold_num}: {fold_accuracy * 100:.2f}%")
    fold_num += 1

# Calculate per-fold average accuracy
print("\n=== Average Accuracy per Fold across all Attack Types ===")
for i, fold_acc_list in enumerate(fold_accuracies, 1):
    avg_fold_acc = np.mean(fold_acc_list)
    print(f"Fold {i} Average Accuracy: {avg_fold_acc * 100:.2f}%")

# Final overall mean accuracy across all folds and attack types
mean_accuracy = np.mean([np.mean(fold) for fold in fold_accuracies])
print("\n=== Stacked Logistic Regression Model (with Logistic Regression, SGD, RFC, DNN) ===")
print(f"Final Mean Accuracy across all folds: {mean_accuracy * 100:.2f}%")
```
